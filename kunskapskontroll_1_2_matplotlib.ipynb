### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# the "object-oriented (OO) style" and "Pyplot style"

# "object-oriented (OO) style" är det rekommenderade sättet, 
# särskilt när du skapar mer komplexa diagram eller hanterar flera figurer och deldiagram. 
# Det är mer explicit och mindre benäget för fel, vilket ger bättre kontroll över olika delar av figuren.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# A figure: Hela fönstret eller ytan där grafer och diagram ritas. 
# Figuren innehåller alla delar av grafen, inklusive axlar, titlar och etiketter.
# Axes: själva grafområdet där data plottas. En figur kan innehålla ett eller flera koordinatsystem. 
# Varje koordinatsystem innehåller en x-axel och en y-axel och kan ha egna etiketter, gränser och stil.
# Axis: Detta refererar till de enskilda x- och y-axlarna i ett koordinatsystem. 
# De styr skalan, gränserna och markeringarna på de horisontella och vertikala sidorna av grafen.
# An artist: Allt man ser i en figur kallas för en artist. 
# Detta inkluderar linjer, text, punkter, titlar, etiketter och även själva figuren och axlarna.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# object-oriented approach
x = np.linspace(-4, 4, 100)
y = x ** 2

fig, ax = plt.subplots()

ax.plot(x, y)

ax.set(title='Graf av funktionen y = x^2', xlabel='x', ylabel='y')


plt.show()

```

```python
# pyplot approach
x = np.linspace(-4, 4, 100)
y = x ** 2

plt.plot(x, y)

plt.title('Graf av funktionen y = x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(random_data_x, random_data_y)
ax1.set(title="Scatter Plot", label="X-axel", ylabel="Y-axel")


ax2.bar(names, values)
ax2.set(title="Bar Plot", xlabel="Fruits", ylabel="Antal")

plt.show()
```
